## Iterate and evaluate your classifier

```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('imdb.txt', sep ="\t", header=None)
df.columns = ['message','good']
df = df.replace(0,'bad').replace(1,'good')
```

```python
keywords = [
    'slow', 'flat', 'disappointed', 'poor', 'crap', 'bad', 'doomed', 'boring',
    'no plot', 'awful', 'sucked', 'horrible', 'terrible', 'average', 'doomed',
    'not funny', 'save', 'wasted', 'negative', 'insipid', 'regret', 'in front',
    'annoying', 'honestly', 'meanders', 'honestly', 'problem', 'horrendous',
    'predictable', 'shallow', 'insincere', 'lousy', 'stupid', 'stand', 'no depth',
    'flawed', 'whatsoever', 'Awful', 'worst', 'disliked', 'appalling', 'too many'
]
for key in keywords:
    df[str(key)] = df.message.str.contains(' ' + str(key) + ' ', case=False)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.heatmap(df.corr(), ax=ax)
```

```python
df['good'] = (df['good'] == 'good')
```

```python
data = df[keywords]
target = df['good']
```

# Classifier 1

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(data, target)
y_pred = bnb.predict(data)
print("Number of bad reviews out of a total of {} reviews: {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(target, y_pred)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=20)
bnb.fit(X_train, y_train).score(X_test, y_test), bnb.fit(data, target).score(data, target)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(bnb, data, target, cv=10)
```

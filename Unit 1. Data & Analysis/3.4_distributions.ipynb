```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Discrete distribution = mass function : only takes #'s
# Continuous distribution = density function
```

```python
rand1 = np.random.normal(50, 300, 1000)
rand2 = np.random.poisson(1, 1000)

rand1.sort()
rand2.sort()

norm = np.random.normal(0, 1, 1000)

norm.sort()
```

```python
plt.plot(norm, rand1, 'o')
plt.show()
```

```python
plt.plot(norm, rand2, "o")
plt.show()
```

```python
plt.hist(rand1, bins=20, color='c')

plt.axvline(rand1.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(rand1.mean() + rand1.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(rand1.mean() - rand1.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
plt.hist(rand2, bins=20, color='c')

plt.axvline(rand2.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(rand2.mean() + rand2.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(rand2.mean() - rand2.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
# p=.5
# probability density function
bernoulli = np.random.binomial(1, .5, 1000)
plt.hist(bernoulli)
plt.show()
```

```python
# n=20 ; p=.5
# probability mass function
binomial = np.random.binomial(20, .5, 1000)
plt.hist(binomial)
plt.show()
```

```python
#shape 5 ; scale 1
#Time until event
# probability density function
gamma = np.random.gamma(5, 1, 1000)
plt.hist(gamma)
plt.show()
```

```python
# lambda = 3  ; rate that events occur in time period.
# Amount something happens within interval
# probability mass function
poisson = np.random.poisson(3, 1000)
plt.hist(poisson)
plt.show()
```

```python
# DataFrame w/ ecommerce data ; amount each customer spent online
# normal distribution ; mean 75 ; sd 25
ecommerce = pd.DataFrame()
ecommerce['spending'] = np.random.normal(75, 25, 1000)
plt.hist(ecommerce['spending'])
plt.show()
```

```python
#subset of ecommerce population
#conditional distribution
ecommerce['visit_count'] = np.random.randint(0, 5, 1000)
#plot customers where visit count is greater than 2
plt.hist(ecommerce[ecommerce['visit_count'] > 2]['spending'])
plt.show()
```

```python
#1
betaset = np.random.beta(10, 10, 100)
plt.hist(betaset)
plt.axvline(betaset.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(betaset.mean() + betaset.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(betaset.mean() - betaset.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
#2
#normal distribution(mean 0, var 1)
#random variables squared and summed
hypothesis_testing = np.random.chisquare(100, 100)
plt.hist(hypothesis_testing)
plt.axvline(hypothesis_testing.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(hypothesis_testing.mean() + hypothesis_testing.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(hypothesis_testing.mean() - hypothesis_testing.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
#3
inefficiency = np.random.pareto(100, 100)
plt.hist(inefficiency)
plt.axvline(inefficiency.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(inefficiency.mean() + inefficiency.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(inefficiency.mean() - inefficiency.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
#4
#amount event happens in interval
calls = np.random.poisson(100, 100)
plt.hist(calls)
plt.axvline(calls.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(calls.mean() + calls.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(calls.mean() - calls.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
#5
# samples from 0 to 1 from power distribution; exp: a-1
powers = np.random.power(100, 100)
plt.hist(powers)
plt.axvline(powers.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(powers.mean() + powers.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(powers.mean() - powers.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
#6
#zipf's law: frequency of item frequency of item inversely prop. to its' rank in frequency table
zeta = np.random.zipf(100, 100)
plt.hist(zeta)
plt.axvline(zeta.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(zeta.mean() + zeta.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(zeta.mean() - zeta.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```
